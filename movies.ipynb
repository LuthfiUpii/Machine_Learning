```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="ticks", color_codes=True)
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['figure.dpi'] = 150
```

```python
df = pd.read_csv("sample_data/movies_metadata.csv")
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df["budget"] = df["budget"].astype(float)
df["popularity"] = df["popularity"].astype(float)
```

```python
df.info()
```

```python
sns.heatmap(df.corr(method ="spearman"), square=True, cmap='mako',annot=True, annot_kws={"fontsize":6})
```

```python
sns.lmplot(x = 'budget', y = 'revenue', data = df)
```

```python
X = df[['budget']]
y = df[['revenue']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```

```python
# predict the y values
y_pred=regressor.predict(X_test)
# a data frame with actual and predicted values of y
evaluate = pd.DataFrame({'Actual': y_test.values.flatten(),'Predicted': y_pred.flatten()})
evaluate.tail(10)
```

```python
evaluate.head(10).plot(kind = 'bar')
```

```python
# Scoring the model
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
# R2 Score
print(f"R2 score: {r2_score(y_test, y_pred)}")
# Mean Squared Error (MSE)
print(f"MSE score: {mean_squared_error(y_test, y_pred)}")
```
